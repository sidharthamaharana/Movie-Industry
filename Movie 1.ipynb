# 4. Data Preprocessing

## 4.1. Data Cleaning

 - Fill missing value

```python
movie["rating"].fillna("R",inplace=True)
```

```python
movie["budget"].fillna(16,inplace=True)
```

```python
movie["company"].fillna("Universal Pictures",inplace=True)
```

```python
movie["country"].fillna("United States",inplace=True)
```

```python
movie["score"].fillna(6.3,inplace=True)
```

```python
movie["votes"].fillna(88106,inplace=True)
```

```python
movie["runtime"].fillna(107,inplace=True)
```

```python
movie.rename(columns={"gross":"revenue"},inplace=True)
```

```python
movie["budget"].replace({16:160000},inplace=True)
```

 - Drop missing value

```python
movie.dropna(inplace=True)
```

 - Wrong datatype

```python
movie["score"]=movie["score"].astype(float)
```

```python
movie["budget"]=movie["budget"].astype(float)
```

```python
movie["revenue"]=movie["revenue"].astype(float)
```

```python
movie["revenue"].fillna(22,inplace=True)
```

```python
movie["votes"]=movie["votes"].astype(float)
```

## 4.2. Feature Selection

```python
movie.drop(["writer","star","director","released"],axis=1,inplace=True)
```

```python
movie["profit"]= (movie["revenue"]-movie["budget"])/movie["budget"]*100
movie
```

## 4.3. Discretization

```python
movie["movie status"]=pd.cut(movie["score"],bins=[0,5,6.5,8.5,10],labels=["worst","average","good","best"])
```
